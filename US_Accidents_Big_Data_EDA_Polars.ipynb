```python
path_file = r"C:\Users\sonid\Downloads\archive (3)\US_Accidents_March23.csv"
parquet_scan = r"C:\Users\sonid\Downloads\archive (3)\US_Accidents_March23.parquet"

data = pl.scan_parquet(parquet_scan).collect()
data = data.with_columns([
    pl.col("End_Lat").fill_null(pl.col("Start_Lat")).cast(pl.Float64),
    pl.col("End_Lng").fill_null(pl.col("Start_Lng")).cast(pl.Float64),
    pl.col("Start_Time").str.to_datetime(),
    pl.col("End_Time").str.to_datetime()
])

analysis_data = data.with_columns([
    pl.col("Start_Time").dt.hour().alias("Start_Hour"),
    pl.col("Start_Time").dt.weekday().alias("Start_Weekday")
])

heatmap_data = (
    analysis_data
    .group_by(["Start_Weekday", "Start_Hour"])
    .count()
    .pivot(index="Start_Weekday", columns="Start_Hour", values="count")
    .sort("Start_Weekday")
)


pd_heatmap = heatmap_data.to_pandas().set_index("Start_Weekday")

plt.figure(figsize=(12, 6))
sns.heatmap(pd_heatmap, cmap="YlOrRd", fmt="d")
plt.title("Accident Hotspots by Day and Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Day of the Week (1=Monday, 7=Sunday)")
plt.show()
```
